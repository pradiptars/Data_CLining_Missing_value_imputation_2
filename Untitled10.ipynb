```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path =r"/content/drive/MyDrive/bengaluru_house_prices.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
missing_values_count = df.isnull().sum()/df.shape[0]*100
missing_values_count
```

```python
missing_col = missing_values_count[missing_values_count>20].keys()
missing_col
```

```python
df2_drop_clm = df.drop(missing_col,axis=1)
df2_drop_clm.shape
```

```python
df3_num = df2_drop_clm.select_dtypes(include=['int64', 'float64'])
df3_num.shape
```

```python
df3_num.isnull().sum()
```

```python
num_ver_miss = ['bath', 'balcony']
df3_num[num_ver_miss][df3_num[num_ver_miss].isnull().any(axis=1)]
```

```python
df['location'].unique()
```

```python
df[df.loc[: 'location'] == 'Electronic City Phase II']["balcony"].replace(np.nan, df[df.loc[: 'location'] == 'Electronic City Phase II']["balcony"].mean())
```

```python
df_copy = df.copy()
for var_class in df['location'].unique():
   df[df.loc[: 'location'] == var_class]["balcony"].replace(np.nan, df[df.loc[: 'location'] == var_class]["balcony"].mean())
```

```python
df_copy.isnull().sum()
```

```python
df_copy = df.copy()
num_ver_miss = ['bath', 'balcony']
cat_ver_miss = ['society', 'size']
for cat_ver_miss, num_ver_miss in zip(cat_ver_miss, num_ver_miss):
  for var_class in df[cat_ver_miss].unique():
    df_copy[df.loc[: cat_ver_miss] == var_class][num_ver_miss].replace(np.nan, df[df.loc[: cat_ver_miss] == var_class][num_ver_miss].mean())
```

```python
df_copy[num_ver_miss].isnull().sum()
```

```python
df_copy = df.copy()
num_ver_miss = ['bath', 'balcony']
cat_ver_miss = ['availability', 'area_type']
for cat_ver_miss, num_ver_miss in zip(cat_ver_miss, num_ver_miss):
  for var_class in df[cat_ver_miss].unique():
    df[df.loc[: cat_ver_miss] == var_class][num_ver_miss].replace(np.nan, df[df.loc[: cat_ver_miss] == var_class][num_ver_miss].mean())
```

```python
df_copy[num_ver_miss].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_ver_miss):
  plt.subplot(2,2,i+1)
  sns.displot(df_copy[var], bins=20, kde_kws={'linewidth':8,'color': 'red'}, Label='mean',)
  sns.distplot(df[var], bins=20, kde_kws={'linewidth':5,'color': 'green'}, Label='original',)
  plt.legend()
```

```python
df_copy_median = df.copy()
num_ver_miss = ['bath', 'balcony']
cat_ver_miss = ['society', 'size']
for cat_ver_miss, num_ver_miss in zip(cat_ver_miss, num_ver_miss):
  for var_class in df[cat_ver_miss].unique():
    df_copy_median[df.loc[: cat_ver_miss] == var_class][num_ver_miss].replace(np.nan, df[df.loc[: cat_ver_miss] == var_class][num_ver_miss].median())
```
